### 4. 編碼偵測及轉換

- 檢測不同語系文字的編碼
- 嘗試將其轉為 ASCII，觀察差異

```python
# !pip install chardet
import chardet
import unicodedata

# Step 1. 先定義兩段文字
text_a = "Café de l’Université"
text_b = "政治學系 Department of Political Science"

# encoding
encode_1 = text_a.encode("utf-8")
encode_2 = text_b.encode("utf-8")

print(f"---- encoding ----\n{encode_1}\n{encode_2}\n")

# 偵測編碼
print(f"\n---- detect ----")
print(chardet.detect(encode_1))
print(chardet.detect(encode_2))

# decoding
decode_1 = encode_1.decode("utf-8")
decode_2 = encode_2.decode("utf-8")

print(f"\n---- decoding ----\n{decode_1}\n{decode_2}\n")

# 正規化、轉為ASCII
ascii_1 = unicodedata.normalize("NFKD", decode_1).encode("ascii", "ignore").decode("ascii")
ascii_2 = unicodedata.normalize("NFKD", decode_2).encode("ascii", "ignore").decode("ascii")

print(f"---- ASCII ----\n{ascii_1}\n{ascii_2}")
```
