```python
# .attrs 屬性（attributes），回傳該 HTML 標籤所有屬性
myLinks = myBS.findAll("a", {"href":re.compile(".*")})
for link in myLinks:
    print(link.attrs)
```

## 樹狀結構

```python
# myBS.h1：找到第一個 <h1> 標籤。
# .children：取得這個 <h1> 底下的「所有子節點」
for myChild in myBS.h1.children:
    print(myChild)
```

```python
# myBS.h2：先找到網頁中的第一個 <h2> 標籤。
# .descendants：取得 所有延伸節點（descendants）：子節點，子子節點
for myDes in myBS.h2.descendants:
    print(myDes)
```

```python
# 找同一層的其他節點
# .next_sibling	下一個平行（只一個）	找 <i> 後面第一個元素
# .next_siblings	所有後面的平行節點	找 <i> 後面全部的元素
for mySib in myBS.find("i").next_siblings:
    print(mySib)
```

```python
# 樹狀結構
print(myBS.prettify())
```

基本的文字前處理 + 字頻統計（word frequency）

```python
### How to write and read a pickle
## One thing you might want to save for later
## is a list of word frequncies

# Create one string with all text from the .txt file I read in
wordstring = ""
for line in lines:
    wordstring = wordstring + str(line)


## 文字處理步驟
# 移除數字與簡單的標點符號
wordstring = re.sub(r"[0-9.,:-]+", "", wordstring)
# 移除某些不能放入正則字元集合的符號（特殊符號），用 alternation（或運算）處理難輸入的符號
wordstring = re.sub(r"(\(|\)|\*|\[|\]|')+", "", wordstring)
# 把換行符號（\n）變成空白
wordstring = re.sub(r"\n", " ", wordstring)
# 把多個空白變成一個空白
wordstring = re.sub(r"\s+", " ", wordstring)
# 全部變成小寫
wordstring = wordstring.lower()
# 用 split() 切成單字列表
wordlist = wordstring.split()

## 建立字頻統計（word frequency dictionary）
# 建立空字典
wordfreq = {}
# 逐字統計：用迴圈讀過每個 wordlist 裡的字
for w in wordlist:
    # 跳過太短的字（例如 “a”, “to”, “of”），常見的 NLP 過濾手法
    if len(w) > 2:
        # 檢查是否是第一次出現
        if w not in wordfreq:
            # 如果是第一次出現，將該字收入字典
            wordfreq[w] = 0
        # 已在字典中，字頻 +1
        wordfreq[w] += 1

## 最後印出字頻統計
print(wordfreq)
```

```python
sorted_freq = sorted(wordfreq.items(), key=lambda x: x[1], reverse=True)
print(sorted_freq[:20])   # 印出前 20 個最常出現的字
```

```python
import csv
import os
from bs4 import BeautifulSoup
import json

mydir = "/content/251107"

# 看看實際有哪些檔案
#print("Files under/content/251107:", os.listdir(mydir))

# 只抓 .json 檔案
dirlist = [f for f in os.listdir(mydir) if f.endswith(".json")]
#print("JSON files:", dirlist)
#print("Number of JSON files:", len(dirlist))

# 如果沒有任何 json 檔，直接提醒
if len(dirlist) == 0:
    print("沒有 json 檔")
else:
    # 建立輸出 CSV 檔
    fout = open("/content/opinions.csv", "w", newline="")
    outfilehandle = csv.writer(fout)
    outfilehandle.writerow(["id", "author", "pages", "wordcount"])

    # 迴圈處理每一個 json
    for entry in dirlist:
        infilepath = mydir + entry
        #print("Processing:", entry)

        infilehandle = json.load(open(infilepath)) #infilehandle 現在是一個字典
        # 抓出三個欄位（用 .get），如果 key 不存在，就回傳你設定的預設值："" 是空字串 以及 或是數字 0
        uID = infilehandle.get('id', "")
        author = infilehandle.get('author', "")
        pages = infilehandle.get('page_count', 0)

        bsOpin = BeautifulSoup(infilehandle.get('html_lawbox', ""), "html.parser") #用 BeautifulSoup 解析 HTML（從 JSON 裡的 html_lawbox）用 BeautifulSoup 解析之後，你就能用 .find() / .findAll() 等方式抓出文字

        opinText = ""
        for para in bsOpin.findAll("p"): #擷取所有 <p> 段落的純文字
            opinText += para.get_text() #para.get_text() → 擷取純文字（不含 HTML 標籤）

        opinWords = opinText.split() #把整篇文字拆成單字
        WC = len(opinWords) #計算字數（word count）

        localrow = [uID, author, pages, WC] #把要寫入 CSV 的一列組起來
        outfilehandle.writerow(localrow) #寫入 CSV 檔案

    fout.close()

```
