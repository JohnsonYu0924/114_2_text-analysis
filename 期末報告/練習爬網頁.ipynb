```python


with open('191218_政見發表會.md', 'r', encoding='utf-8') as t:
    transcript = t.read()
    print(type(transcript)) #class

senten = []
transcript = re.split(r"\n", transcript)
for i in transcript:
    if i != "":
        senten.append(i.strip())
# print(senten)

candidate = input("要查詢的人：")
candidate = '#### '+ candidate
candidateTranscript = []
Flag = False
for i in senten:
    if i == candidate:
        Flag = True
        print(i)
    if Flag and not re.search(r"#### ", i):
        candidateTranscript.append(i)
print(candidateTranscript[-1])
# for i in candidateTranscript:
#     print(i)
```
