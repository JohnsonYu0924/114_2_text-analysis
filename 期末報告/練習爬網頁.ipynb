### 第五題：從 JSON 的 html_lawbox 中取出文字並計算字數

````python
# 你的程式碼
bsOpin = BeautifulSoup(opJs.get("html_lawbox", ""), "html.parser")
# print(bsOpin)

lawBox = ""
for para in bsOpin.find_all("p"):
  lawBox = lawBox + " " + para.get_text()
print(f"---- 去除<p> ----\n{lawBox}")

lawBox = re.sub(r"[0-9.,:-]+", "", lawBox)
lawBox = re.sub(r"(\(|\)|\*|\[|\[|\]|'|\"|\+)+", "", lawBox)
lawBox = re.sub(r"\n", " ", lawBox)
lawBox = lawBox.lower()
print(f"\n---- 清除和小寫 ----\n{lawBox}")

textList = lawBox.split()
print(f"\n---- 拆成list ----\n{textList}")
print(f"\n---- 計算總字數 ----\n{len(textList)}")


# 感覺可以繼續往下洗，像是有出現'e', 'a', 's'這種無意義字？
print("\n\n==== 繼續往下洗 ====")
nTextList = []
for text in textList:
  if len(text) > 1:
    nTextList.append(text)
print(f"\n---- 洗掉單一字母 ----\n{nTextList}")
print(f"\n---- 計算總字數 ----\n{len(nTextList)}")
````

### 第六題：從 TXT 檔案讀取 OPINION 段落

可以先上傳 myExample.txt 到 Colab



````python
# 你的程式碼
txt = open("/content/drive/MyDrive/1117作業/myExample.txt", "r", encoding="utf8").readlines()
# for t in txt:
#   print(t)

opinText = ""
flag = False
for line in txt:
  if re.search(r"^[\s]*OPINION$", line):
    flag = True
  if flag and not re.search(r"^[\s]*$", line):
    opinText += line

opinText = re.sub(r"\s+", " ", opinText)
print(opinText)
````

### 第七題：批次處理資料夾中的 JSON 並輸出 CSV

- id
- author
- page_count
- html_lawbox

````python
# 你的程式碼
path = "/content/drive/MyDrive/1117作業"
!ls "/content/drive/MyDrive/1117作業"

os.chdir(path)
print(os.getcwd())

dirList = [i for i in os.listdir(path) if i.endswith(".json")]
if len(dirList) == 0:
  print("There's no '.json' data.")
else:
  # 建立.csv和欄位設定
  font = open(path+"/opinions.csv", "w", newline="")
  outFile = csv.writer(font,
                     delimiter=",",
                     quotechar=" ",
                     quoting=csv.QUOTE_NONNUMERIC)
  localRow = ["ID", "Author", "PageCount", "lawBox"]
  outFile.writerow(localRow)

  #處理.json檔
  for i in dirList:
    ipath = path + "/" + i
    inFile= json.load(open(ipath))

    #先處理要放進欄位的資料(特別是lawbox)
    bsOpin = BeautifulSoup(inFile.get("html_lawbox", ""), "html.parser")

    lawBox = ""
    for para in bsOpin.find_all("p"):
      lawBox = lawBox + " " + para.get_text()
    lawBox = re.sub(r"[0-9.,:-]+", "", lawBox)
    lawBox = re.sub(r"(\(|\)|\*|\[|\[|\]|'|\"|\+)+", "", lawBox)
    lawBox = re.sub(r"\n", " ", lawBox)
    lawBox = lawBox.lower()
    print(f"{i}\n{lawBox}")

    ID = inFile.get("id", "")
    author = inFile.get("author", "")
    pageCount = inFile.get("page_count", 0)
    #lawBox上面處理完了

    #寫檔
    localRowes = [ID, author, pageCount, lawBox]
    print(localRowes)
    outFile.writerow(localRowes)

  font.close()
````
